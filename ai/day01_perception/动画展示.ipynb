```python
## 
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("data-ganzhiji.csv",delimiter=',')
X = data[:,0:2]
y = data[:,2:3]

##  wb
W = np.array(np.random.rand(2,1))
b = np.random.rand(1)[0]

## 
def stepFunction(value):
    if value>=0:
        return 1  #
    else :
        return 0  #
    
    
## w1*x1 + w2*x2+b -->stepFunction
## 
def perception(X,W,b):
    return stepFunction((np.matmul(X,W)+b)[0])
learn_rate = 0.01

fig = plt.figure(figsize=(5,5),dpi=80)
for epoch in range(100):
        ## 
    for i in range(len(X)):
        y_hat = perception(X[i],W,b) #
        y_real = y[i] ## 
        if y_real - y_hat == 1: #10 
            W[0] += X[i][0]*learn_rate
            W[1] += X[i][1]*learn_rate
            b += learn_rate*1
        if y_real - y_hat == -1: #01 
            W[0] -= X[i][0]*learn_rate
            W[1] -= X[i][1]*learn_rate
            b -= learn_rate*1
        if y_real- y_hat == 0: ## 11 11
            pass

    for i in range(len(y)):
        if(y[i]==0): #
            plt.scatter(X[i][0],X[i][1],c='y')
        else:
            plt.scatter(X[i][0],X[i][1],c='r')

    XX = np.linspace(-2,2,10)
    Y = -(b+W[0]*XX)/W[1]
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    plt.plot(XX,Y)
    plt.show()
```
